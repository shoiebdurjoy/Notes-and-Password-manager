```python
class User:
    def __init__(self, username, email, master_password):
        self.username = username
        self.email = email
        self.master_password = master_password

class Manager:
    users = {}  # Class variable to store users {email: User}

    @classmethod
    def sign_up(cls, username, email, master_password):
        if email in cls.users:
            return False
        cls.users[email] = User(username, email, master_password)
        return True

    @classmethod
    def sign_in(cls, email, master_password):
        if email in cls.users and cls.users[email].master_password == master_password:
            return cls.users[email]
        return None

class ItemManager:
    def __init__(self, user):
        self.user = user
        self.items = {}  # {title: {'doc': doc, 'content': content}}

    def add_item(self, title, doc, content):
        self.items[title] = {'doc': doc, 'content': content}

class PasswordManager(ItemManager):
    def __init__(self, user):
        super().__init__(user)

    def add_password(self, title, password):
        self.add_item(title, '', password)

    def get_password_info(self, title):
        return self.items.get(title, "Password not found.")

class NoteManager(ItemManager):
    def __init__(self, user):
        super().__init__(user)

    def add_note(self, title, doc, content):
        self.add_item(title, doc, content)

    def update_title(self, title, new_title):
        if title in self.items:
            item = self.items.pop(title)
            self.items[new_title] = item

manager = Manager()

while True:
    print("Welcome to the Password and Note Manager!")
    choice = input("Enter '1' to sign in or '2' to sign up: ")

    if choice == '1':
        email = input("Email: ")
        master_password = input("Master Password: ")

        active_user = manager.sign_in(email, master_password)
        if active_user:
            print("Sign in successful!")
            password_manager = PasswordManager(active_user)
            note_manager = NoteManager(active_user)
            break
        else:
            print("Invalid email or master password.")

    elif choice == '2':
        username = input("Username: ")
        email = input("Email: ")
        master_password = input("Master Password: ")

        if manager.sign_up(username, email, master_password):
            print("Sign up successful! You can now sign in.")
        else:
            print("Email is already taken. Please choose another.")

    else:
        print("Invalid choice.")

while True:
    print("\n1. Password Manager")
    print("2. Note Manager")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        print("\n1. Add Password")
        print("2. Get Password Info")
        print("3. Update Username")
        print("4. Update Email")
        print("5. Update Password")

        password_choice = input("Enter your password choice: ")

        if password_choice == '1':
            title = input("Enter password title: ")
            password = input("Enter password: ")
            password_manager.add_password(title, password)
            print("Password added successfully!")

        elif password_choice == '2':
            title = input("Enter password title: ")
            password_info = password_manager.get_password_info(title)
            print("Password:", password_info)

        elif password_choice == '3':
            new_username = input("Enter new username: ")
            manager.update_username(active_user, new_username)
            print("Username updated successfully!")

        elif password_choice == '4':
            new_email = input("Enter new email: ")
            manager.update_email(active_user, new_email)
            print("Email updated successfully!")

        elif password_choice == '5':
            new_password = input("Enter new password: ")
            manager.update_password(active_user, new_password)
            print("Password updated successfully!")

        else:
            print("Invalid choice.")

    elif choice == '2':
        print("\n1. Add Note")
        print("2. Update Title")
        print("3. Update Doc")
        print("4. Update Content")
        print("5. Get Content")
        print("6. Get Doc")
        print("7. Finish")

        note_choice = input("Enter your note choice: ")

        if note_choice == '1':
            title = input("Enter note title: ")
            doc = input("Enter doc: ")
            content = input("Enter content: ")
            note_manager.add_note(title, doc, content)
            print("Note added successfully!")

        elif note_choice == '2':
            title = input("Enter note title: ")
            new_title = input("Enter new title: ")
            note_manager.update_title(title, new_title)
            print("Title updated successfully!")

        elif note_choice == '3':
            title = input("Enter note title: ")
            new_doc = input("Enter new doc: ")
            note_manager.update_doc(title, new_doc)
            print("Doc updated successfully!")

        elif note_choice == '4':
            title = input("Enter note title: ")
            new_content = input("Enter new content: ")
            note_manager.update_content(title, new_content)
            print("Content updated successfully!")

        elif note_choice == '5':
            title = input("Enter note title: ")
            content = note_manager.get_content(title)
            print("Content:", content)

        elif note_choice == '6':
            title = input("Enter note title: ")
            doc = note_manager.get_doc(title)
            print("Doc:", doc)

        elif note_choice == '7':
            print("Returning to main menu.")

        else:
            print("Invalid choice.")

    elif choice == '3':
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")            
```
